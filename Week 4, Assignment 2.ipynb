```python
import numpy as np
import pandas as pd
```

```python
years = ['2021-12-31', '2022-12-31', '2023-12-31', '2024-12-31', '2025-12-31', '2026-12-31', '2027-12-31', '2028-12-31', '2029-12-31', '2030-12-31', '2031-12-31', '2032-12-31']
Revenue = pd.Series(index=years, dtype ='float64')
Revenue['2021-12-31'] = 222  
Revenue
```

```python
growth_rate = 0.18
for year in range(1, 12):
    Revenue[year] = Revenue[year - 1] * (1 + growth_rate)  
ebitda_margin = 0.45
EBITDA = Revenue * ebitda_margin
tax_rate = 0.21
Tax = -EBITDA * tax_rate 
Capital_Expenditures_rate = 0.03
Capital_Expenditures = -Capital_Expenditures_rate * Revenue 
Working_Capital_rate = 0.06
Working_Capital = -Working_Capital_rate * Revenue 
Cash_Flow = EBITDA + Tax + Capital_Expenditures + Working_Capital
Discount_rate = 0.06
Discount_factor = pd.Series(index = years, dtype='float64')
for year in range(0,12):
    Discount_factor[year]= (1 - Discount_rate) ** year
Present_Value_FCF = (Discount_factor * Cash_Flow)
```

```python
output = pd.DataFrame(index = years , columns =['Revenue', 'EBITDA', 'Tax', 'Capital_Expenditures', 'Working_Capital', 'Cash_Flow', 'Discount_Factor', 'Present_Value_FCF'])
output['Revenue']=Revenue
output['EBITDA']= EBITDA
output['Tax']= Tax
output['Capital_Expenditures']=Capital_Expenditures
output['Working_Capital']=Working_Capital
output['Cash_Flow'] = Cash_Flow
output['Discount_Factor']= Discount_factor
output['Present_Value_FCF']=Present_Value_FCF
output
```

```python
pv = output.Present_Value_FCF[1:-1].sum()
terminal_value = 32.7 * output.EBITDA[-1]
present_value_of_terminal_value = terminal_value * output.Discount_Factor[-2]
present_value_of_asset = pv + present_value_of_terminal_value
print('Our projected Present Value of Projected Future Cash Flows are:', str(output.Present_Value_FCF[1:-1].sum()) )
print('We are providing revenue guidance in 2032 to be:',  str(Revenue[11]) )
print('With an estimated EBITDA around:', str(EBITDA[11]) )
print('The terminal value at this time given a multiple of 32.7 and the above EBITDA for 2032, would equal:' +  str(present_value_of_terminal_value))
print('The present value of asset:', str(present_value_of_asset))
```
